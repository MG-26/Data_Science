```python
# Create the dataframe df_by_genre

# Create the dataframe df_by_genre
df_by_genre = movies.iloc[:,11:60]

# Dropping the columns not required
df_by_genre.drop(['MetaCritic','Runtime'], axis = 1,inplace = True)


```

```python
# Checking the number of columns
df_by_genre.shape
```

```python
# Create a column cnt and initialize it to 1

df_by_genre['cnt'] = 1
```

```python
df_by_genre.head()
```

```python
# Group the movies by individual genres
df_by_genre1 = df_by_genre.groupby('genre_1').agg(np.sum)
df_by_genre2 = df_by_genre.groupby('genre_2').agg(np.sum)
df_by_genre3 = df_by_genre.groupby('genre_3').agg(np.sum)
```

```python
df_by_genre1
```

```python
df_by_genre1.loc["Action","cnt"]
```

```python
# Add the grouped data frames and store it in a new data frame

df_by_genre12=df_by_genre1.add(df_by_genre2,fill_value=0)
df_add=df_by_genre12.add(df_by_genre3,fill_value=0)
df_add.head()
```

```python
# Extract genres with atleast 10 occurences
top=df_add.sort_values(by="cnt", ascending=False)
top.head(10)
```

```python
genre_top10=df_add[df_add.cnt>=10]
genre_top10.iloc[:10]
```

```python
genre_top10.shape
```

```python
genre_top10.cnt.min()
```

```python
genre_top10.columns
```

```python
# Take the mean for every column by dividing with cnt 
for i in genre_top10.columns:
    if i != 'cnt':
        genre_top10[i] = genre_top10[i]/genre_top10['cnt']
    else:
        break

genre_top10
```

```python
# Rounding off the columns of Votes to two decimals

genre_top10=genre_top10.apply(lambda x: round(x,2),axis=0)
genre_top10
```

```python
# Converting CVotes to int type

genre_top10[['CVotes10', 'CVotes09', 'CVotes08', 'CVotes07', 'CVotes06', 'CVotes05',
       'CVotes04', 'CVotes03', 'CVotes02', 'CVotes01', 'CVotesMale',
       'CVotesFemale', 'CVotesU18', 'CVotesU18M', 'CVotesU18F', 'CVotes1829',
       'CVotes1829M', 'CVotes1829F', 'CVotes3044', 'CVotes3044M',
       'CVotes3044F', 'CVotes45A', 'CVotes45AM', 'CVotes45AF', 'CVotes1000',
       'CVotesUS', 'CVotesnUS']] = genre_top10[['CVotes10', 'CVotes09', 'CVotes08', 'CVotes07', 'CVotes06', 'CVotes05',
       'CVotes04', 'CVotes03', 'CVotes02', 'CVotes01', 'CVotesMale',
       'CVotesFemale', 'CVotesU18', 'CVotesU18M', 'CVotesU18F', 'CVotes1829',
       'CVotes1829M', 'CVotes1829F', 'CVotes3044', 'CVotes3044M',
       'CVotes3044F', 'CVotes45A', 'CVotes45AM', 'CVotes45AF', 'CVotes1000',
       'CVotesUS', 'CVotesnUS']].astype(int)
```
